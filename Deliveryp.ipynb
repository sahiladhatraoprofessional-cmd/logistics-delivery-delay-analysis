```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Admin\Downloads\logistics_delivery_dataset.csv")
```

```python
df.head()
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Ship_Date"] = pd.to_datetime(df["Ship_Date"])
df["Delivery_Date"] = pd.to_datetime(df["Delivery_Date"])
```

```python
df["delivery_days"] = (df["Delivery_Date"] - df["Ship_Date"]).dt.days
```

```python
df.groupby("Region")["delivery_days"].mean()
```

```python
df.groupby("Warehouse")["delivery_days"].mean()
```

```python
df.groupby("Shipping_Mode")["delivery_days"].mean()
```
